# Feature Selection

```python
movies.columns
```

```python
movies = movies[['id', 'title', 'genres', 'overview']]
```

```python
movies
```

```python
m_genre = movies['genres']
```

```python
m_genre
```

```python
movies.loc[:, 'tags'] = movies['genres'] + movies['overview']
```

```python
movies
```

```python
new_data = movies.drop(columns=['genres', 'overview'])
```

```python
new_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=4803, stop_words='english')
```

```python
cv
```

```python
vector = cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```
